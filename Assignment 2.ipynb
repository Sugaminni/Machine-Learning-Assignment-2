```python
# Imports

import numpy as np
import pandas as pd

from sklearn.datasets import fetch_california_housing
from sklearn.datasets import load_breast_cancer
```

```python
# Linear Regression And Data Loading

california = fetch_california_housing(as_frame=True)
df = california.frame
print(df.head())

# Features and target
X = df.drop("MedHouseVal", axis=1).values
y = df["MedHouseVal"].values

# Adds Bias
X = np.c_[np.ones(X.shape[0]), X]

# 70/30 Split
split = int(0.7 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
# Mean Squared Error function
def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Initializes Params
theta = np.zeros(X_train.shape[1])

learning_rate = 0.01
iterations = 1000
m = len(y_train)

# Gradient Descent
for _ in range(iterations):
    predictions = X_train.dot(theta)
    errors = predictions - y_train

    gradient = (1/m) * X_train.T.dot(errors)
    theta -= learning_rate * gradient

# Predictions and Eval
test_predictions = X_test.dot(theta)
mse = mean_squared_error(y_test, test_predictions)
print("MSE:", mse)
```

```python
# Logistic Regression And Data Loading

data = load_breast_cancer()
X = data.data
y = data.target

# Adds Bias
X = np.c_[np.ones(X.shape[0]), X]

# 70/30 Split
split = int(0.7 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Initializes
theta = np.zeros(X_train.shape[1])
learning_rate = 0.01
iterations = 1000
m = len(y_train)

# Gradient Descent
for _ in range(iterations):
    z = X_train.dot(theta)
    predictions = sigmoid(z)
    errors = predictions - y_train
    gradient = (1/m) * X_train.T.dot(errors)
    theta -= learning_rate * gradient

# Eval
z_test = X_test.dot(theta)
probabilities = sigmoid(z_test)
predicted_classes = (probabilities >= 0.5).astype(int)
accuracy = np.mean(predicted_classes == y_test)
print("Accuracy:", accuracy)
```
